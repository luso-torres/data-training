```python
import csv

def process_csv(input_file, output_file):
    with open(input_file, 'r', newline='') as infile:
        reader = csv.reader(infile)
        rows = [row for row in reader]

    with open(output_file, 'w', newline='') as outfile:
        writer = csv.writer(outfile)
        for row in rows:
            new_row = []
            for item in row:
                new_row.extend(item.split(','))
            writer.writerow(new_row)

# Example usage
input_file = 'output.csv'
output_file = 'output2.csv'
process_csv(input_file, output_file)
```

```python
import csv

def process_csv(input_file, output_file):
    with open(input_file, 'r', newline='') as infile:
        reader = csv.reader(infile)
        rows = [row for row in reader]

    with open(output_file, 'w', newline='') as outfile:
        writer = csv.writer(outfile)
        for row in rows:
            new_row = []
            for i, item in enumerate(row):
                # Split the item based on comma
                split_items = item.split(',')
                if i == len(row) - 1:
                    # Handle the last item separately
                    new_row.extend(split_items)
                else:
                    new_row.extend(split_items[:-1])
                    new_row.append(split_items[-1])
            writer.writerow(new_row)

# Example usage
input_file = 'output.csv'
output_file = 'output3.csv'
process_csv(input_file, output_file)

```

```python
import csv

def process_csv(input_file, output_file):
    with open(input_file, 'r', newline='') as infile:
        reader = csv.reader(infile)
        rows = [row for row in reader]

    with open(output_file, 'w', newline='') as outfile:
        writer = csv.writer(outfile)
        for row in rows:
            if row:
                *rest, last_column = row
                last_digit = last_column.rsplit(',', 1)[-1]
                new_row = rest + [last_column[:-len(last_digit)-1], last_digit]
                writer.writerow(new_row)

# Example usage
input_file = 'output.csv'
output_file = 'Xddd.csv'
process_csv(input_file, output_file)
```

```python
import csv

def process_csv(input_file, output_file):
    with open(input_file, 'r', newline='') as infile:
        reader = csv.reader(infile)
        rows = [row for row in reader]

    with open(output_file, 'w', newline='') as outfile:
        writer = csv.writer(outfile)
        i=0
        for row in rows:
            if i == 0:
                new_row = row
                writer.writerow(new_row)
            elif row:
                # Remove everything before the single digit in the last column
                    *rest, last_column = row
                    last_digit = last_column[-1]  # Assumes the last character is the single digit
                    new_row = rest + [last_digit]
                    writer.writerow(new_row)
            i+=1

# Example usage
input_file = 'Xddd.csv'
output_file = 'clean_file.csv'
process_csv(input_file, output_file)
```

```python
import csv

def process_csv(input_file, output_file):
    with open(input_file, 'r', newline='') as infile:
        reader = csv.reader(infile)
        rows = [row for row in reader]

    with open(output_file, 'w', newline='') as outfile:
        writer = csv.writer(outfile)
        writer.writerow(rows[0])  # Write the first row as is

        for row in rows[1:]:
            if row:
                # Remove everything before the single digit in the last column
                *rest, last_column = row
                last_digit = last_column[-1]  # Assumes the last character is the single digit
                new_row = rest + [last_digit]
                writer.writerow(new_row)

# Example usage
input_file = 'output.csv'
output_file = 'clean_output.csv'
process_csv(input_file, output_file)
```

```python
import csv
import re

def process_csv(input_file, output_file):
    with open(input_file, 'r', newline='') as infile:
        reader = csv.reader(infile)
        rows = [row for row in reader]

    with open(output_file, 'w', newline='') as outfile:
        writer = csv.writer(outfile)
        writer.writerow(rows[0] + ["Split_1", "Split_2"])  # Write the first row with additional column headers

        for row in rows[1:]:
            if row:
                *rest, last_column = row
                match = re.search(r'(\D)(\d+)', last_column)
                if match:
                    new_row = rest + [last_column[:match.start(1)], match.group(2)]
                else:
                    new_row = rest + [last_column, ""]
                writer.writerow(new_row)

# Example usage
input_file = 'output.csv'
output_file = 'right.csv'
process_csv(input_file, output_file)
```
